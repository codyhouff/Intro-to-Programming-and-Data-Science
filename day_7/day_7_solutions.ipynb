Load the `diamonds` dataset from `sns` and do the same thing: come up with a question about it and create a visualization that helps to answer it.

```python
# Is price correlated with carat, and is that correlation the same for every cut? 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.lmplot(x='carat', 
           y='price', 
           col = 'cut', 
           col_wrap= 2,
           order = 2, 
           data=diamonds, 
           line_kws ={'color':'red'}, 
           hue ='cut')
```

```python
sns.relplot(kind='scatter', x = 'carat', y = 'price', data = diamonds,
           alpha = .02, size = 'depth')
```

```python
ax = sns.lmplot(x = 'depth', y = "price", hue = "clarity", col = "cut", data = diamonds);
```
